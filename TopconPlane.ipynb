```python
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
dataSet = pd.read_csv("plane.csv")
dataSet.head(5)
```

```python
numArray = dataSet.to_numpy()

data = numArray.tolist()

listx = []
listy = []
listz = []
for i,j,k in data:
    listx.append(i)
    listy.append(j)
    listz.append(k)

x = np.array(listx)
y = np.array(listy)
z = np.array(listz)
```

```python
tmp_A = []
tmp_b = []
for i in range(len(x)):
    tmp_A.append([x[i], y[i], 1])
    tmp_b.append(z[i])
b = np.matrix(tmp_b).T
A = np.matrix(tmp_A)

fit = (A.T * A).I * A.T * b
```

```python
a = (float)(fit[0])
b = (float)(fit[1])
c = -1
d = (float)(fit[2])
```

```python
x1 = np.linspace(-1,0.3,10)
y1 = np.linspace(0.8,1.3,10)

X,Y = np.meshgrid(x1,y1)
Z=a*X + b*Y + d
```

```python
fig = plt.figure()

ax = fig.add_subplot(111,projection='3d')
ax.set_xlim(-0.8, 0.2)
ax.set_ylim(0.85, 1.25)
ax.set_zlim(-0.8, 0.3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.scatter3D(x,y,z)

ax.plot_surface(X, Y, Z, alpha=0.2)
print("%f x + %f y - 1z + %f = 0" % (fit[0], fit[1], fit[2]))
```
